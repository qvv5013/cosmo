```python
def compute_Ree(traj):
    n_frames=traj.xyz.shape[0]
    Ree=np.zeros(n_frames)
    for f in range(n_frames):
        coord1 = traj.xyz[f,0]
        coord2 = traj.xyz[f,-1]
        Ree[f] = np.linalg.norm(coord2-coord1)
        
    return Ree
```

```python
Ree = compute_Ree(traj)
```
